```python
# Calculate Features
AUC = []
PEAK_PEAK = []
Mean = []

for i in range(x_mag.shape[0]):
    signal = Actigraph_Metrics.VecMag_MAD(x_mag[i,:],100)
    x = []
    for i in range(len(signal)):
        x.append(i)
    Mean.append(np.mean(signal))
    AUC.append(Actigraph_Metrics.calc_area_under_curve(x,signal))
    PEAK_PEAK.append(max(signal)-min(signal))
    
print(len(x_mag))
```

```python
# Spectral Skewness Calculation

x_skew = []
for data_point in x_mag:
    # Compute the spectrum (you need to define how you calculate it based on your data)
    spectrum = np.fft.fft(data_point)
    
    mean_spectrum = np.mean(spectrum)
    std_spectrum = np.std(spectrum)
    skewness_spectrum = np.mean(((spectrum - mean_spectrum) / std_spectrum) ** 3)
    
    # Append the skewness value to x_skew array
    x_skew.append(skewness_spectrum)
```

```python
# Graph Spectral Skewness vs. SBS

plt.figure(figsize=(10, 6))  # Adjust figure size as needed

# Plot mean against time
plt.scatter(SBS[0], x_skew, color='green')

plt.xticks([-1, 0, 1, 2])
# Labeling the axes
plt.xlabel('SBS')
plt.ylabel('Spectral Skewness')

# Title for the plot
plt.title('SBS and Spectral Skewness')

# Display the plot
plt.show()
```
